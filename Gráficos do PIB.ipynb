```python
import pandas as pd
import altair as alt
import numpy as np
```

```python
dates = ['2018-11-30', '2019-02-28', '2019-05-30', '2019-08-29', '2019-12-03', 
         '2020-03-04', '2020-05-29', '2020-09-01', '2020-12-03']

values = [0, 0.1, -0.2, 0.4, 0.6, 0.5, -1.5, -9.7, 7.7]

df = pd.DataFrame({
    'datas': dates,
    'Variação do PIB': values
})

df['datas'] = pd.to_datetime(df['datas'])
df['colors'] = df['Variação do PIB'].apply(lambda x: x > 0)
```

```python
domain_pd = pd.to_datetime(['2018-11-30', '2020-12-30']).astype(int) / 10 ** 6

alt.Chart(df).mark_bar(size=20).encode(
    x=alt.X('datas', timeUnit='yearmonthdate', scale=alt.Scale(domain=list(domain_pd))),
    y='Variação do PIB',
    color = alt.Color('colors', scale=alt.Scale(range=["#ae1325", "#06982d"]), legend=None),
    tooltip=['datas', 'Variação do PIB']
).properties(width=400)
```

```python
domain_pd = pd.to_datetime(['2018-11-30', '2020-12-30']).astype(int) / 10 ** 6

alt.Chart(df).mark_line().encode(
    x=alt.X('datas', timeUnit='yearmonthdate', scale=alt.Scale(domain=list(domain_pd))),
    y='Variação do PIB',
    tooltip=['datas', 'Variação do PIB']
).properties(width=400)
```

```python
df.at[0, 'Variação do PIB'] = 100.0
df['pib'] = df['Variação do PIB'].expanding(2).sum()
df = df.fillna(100.0)

domain_pd = pd.to_datetime(['2018-11-30', '2020-12-30']).astype(int) / 10 ** 6

alt.Chart(df).mark_line().encode(
    x=alt.X('datas', timeUnit='yearmonthdate', scale=alt.Scale(domain=list(domain_pd))),
    y=alt.Y('pib', title='Percentual do PIB em relação ao 4º trim. de 2018', scale=alt.Scale(domain=[90,110])),
    tooltip=['datas', 'Variação do PIB']
).properties(width=400)
```

```python
a = alt.Chart(df).mark_bar(size=20).encode(
    x=alt.X('datas', timeUnit='yearmonthdate', scale=alt.Scale(domain=list(domain_pd))),
    y='Variação do PIB',
    color = alt.Color('colors', scale=alt.Scale(range=["#ae1325", "#06982d"]), legend=None),
    tooltip=['datas', 'Variação do PIB']
)

b = alt.Chart(df).mark_line().encode(
    x=alt.X('datas', timeUnit='yearmonthdate', scale=alt.Scale(domain=list(domain_pd))),
    y='Variação do PIB',
    tooltip=['datas', 'Variação do PIB']
)

a+b
```

```python
a = alt.Chart(df).mark_text(dx=17, dy=-5).encode(
    x=alt.X('datas', timeUnit='yearmonthdate', scale=alt.Scale(domain=list(domain_pd))),
    y=alt.Y('Variação do PIB', axis=None, scale=alt.Scale(domain=[-12, 10])),
    color = alt.Color('colors', scale=alt.Scale(range=["#ae1325", "#06982d"]), legend=None),
    text=alt.Y('Variação do PIB'),
    tooltip=['datas', 'Variação do PIB']
)


b = alt.Chart(df).mark_line().encode(
    x=alt.X('datas', timeUnit='yearmonthdate', scale=alt.Scale(domain=list(domain_pd))),
    y=alt.Y('Variação do PIB', axis=None),
    tooltip=['datas', 'Variação do PIB']
)

(a + b).configure_axis(
    grid=False
)
```
